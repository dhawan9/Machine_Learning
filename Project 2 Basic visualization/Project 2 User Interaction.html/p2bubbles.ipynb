```python
import pandas as pd
import numpy as np
factbook = pd.read_csv("Datasets/factbook.csv")
for c in factbook.columns:
    if factbook[c].dtype == object:
        factbook[c]=factbook[c].str.replace("[$,)]", "")
        factbook[c]=factbook[c].str.replace("(","-")
        try:

            factbook[c]=factbook[c].astype(float)
        except:
            pass
        

        
factbook.shape
```

```python
factbook
```

```python
from bokeh.io import curdoc, show
from bokeh.models import ColumnDataSource, Grid, LinearAxis, Plot, Scatter

source = ColumnDataSource(factbook.to_dict(orient="list"))

plot = Plot(
    title=None, width=300, height=300,
    min_border=0, toolbar_location=None)

glyph = Scatter(x="GDP per Capita", y="Life expectancy", size=20, fill_color="#74add1")
plot.add_glyph(source, glyph)

xaxis = LinearAxis()
plot.add_layout(xaxis, 'below')

yaxis = LinearAxis()
plot.add_layout(yaxis, 'left')

plot.add_layout(Grid(dimension=0, ticker=xaxis.ticker))
plot.add_layout(Grid(dimension=1, ticker=yaxis.ticker))

curdoc().add_root(plot)

show(plot)
```
